## Problem 1. Fibonacci [2 points]
Given `n`, implement function `fib(n)` that calculates `n`-th Fibonacci number. <br>
Assume `n` is a positive integer and `fib(0)`=0, `fib(1)`=1.

```python
def fib(n):
    # BEGIN_YOUR_CODE
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)
    # END_YOUR_CODE
```

```python
for i in range(10):
    print(f"fib({i}) returns {fib(i)}")
```

## Problem 2. Maximum Word Count [2 points]

You should open a file whose name is `filename` and return two things as a tuple: <br>
The first one is a set of words that occur the maximum number of times and the second one is their counts.

```python
def max_word_count(filename):
    # BEGIN_YOUR_CODE
    word_dict = {}
    non_lang_expressions = ["'", '"', '.', ',', '!', '?']
    with open(filename, 'r') as f:
        while True:
            line = f.readline()
            if not line:
                break
            line = line.split(' ')
            for idx, word in enumerate(line):
                # remove non alphabet expressions such as '?', '!' from word
                _idx = 0
                while _idx < len(word):
                    if word[_idx] in non_lang_expressions:
                        del word[_idx]
                    else:
                        _idx += 1
                # add word to dictionary
                if word in word_dict.keys():
                    word_dict[word] += 1
                else:
                    word_dict[word] = 1
    word_dict["*****"] = -100
    # count max word
    key_list = word_dict.keys()
    if len(key_list) == 0:
        raise ValueError("txt file is empty!")
    max_word = ["*****"]
    for key in key_list:
        if word_dict[key] > word_dict[max_word[0]]:
            max_word = [key]
        elif word_dict[key] == word_dict[max_word[0]]:
            max_word.append(key)
    return max_word, word_dict[max_word[0]]
                
            
    #raise Exception("Not implemented yet")
    # END_YOUR_CODE
```

## Problem 3. Average Price [3 points]
Given `cars`, implement function `average_prices(cars)` that returns a dictionary which contains each `brand` as a key and the average price of cars of that `brand` as a value.

```python
class car:
    def __init__(self, args):
        brand, model, price =  args
        self.brand = brand
        self.model = model
        self.price = price

def average_prices(cars): 
    # BEGIN_YOUR_CODE
    out = {}
    car_num_per_brand = {}
    for car in cars:
        if car.brand not in out:
            out[car.brand] = car.price
            car_num_per_brand[car.brand] = 1
        else:
            out[car.brand] += car.price
            car_num_per_brand[car.brand] += 1
    for brand in out.keys():
        out[brand] /= car_num_per_brand[brand]
    return out
    raise Exception("Not implemented yet")
    # END_YOUR_CODE
```
