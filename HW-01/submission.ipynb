## Problem 1. Fibonacci [2 points]
Given `n`, implement function `fib(n)` that calculates `n`-th Fibonacci number. <br>
Assume `n` is a positive integer and `fib(0)`=0, `fib(1)`=1.

```python
def fib(n):
    # BEGIN_YOUR_CODE
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)
    # END_YOUR_CODE
```

## Problem 3. Average Price [3 points]
Given `cars`, implement function `average_prices(cars)` that returns a dictionary which contains each `brand` as a key and the average price of cars of that `brand` as a value.

```python
class car:
    def __init__(self, args):
        brand, model, price =  args
        self.brand = brand
        self.model = model
        self.price = price

def average_prices(cars): 
    # BEGIN_YOUR_CODE
    out = {}
    car_num_per_brand = {}
    for car in cars:
        if car.brand not in out:
            out[car.brand] = car.price
            car_num_per_brand[car.brand] = 1
        else:
            out[car.brand] += car.price
            car_num_per_brand[car.brand] += 1
    for brand in out.keys():
        out[brand] /= car_num_per_brand[brand]
    return out
    # END_YOUR_CODE
```

## Problem 4. Euclidean Distance [2 points]
Given two numpy arrays `arr1` and `arr2`, implement function `manhattan_distance` that calculates Manhattan distance between `arr1` and `arr2`. <br>
The answer should be rounded to two decimal places. <br>
You need to utilize numpy library for this problem.

```python
def manhattan_distance(arr1, arr2):
    # BEGIN_YOUR_CODE
    return round(sum(abs(arr1 - arr2)), 2)
    
    # END_YOUR_CODE
```

3. The modified csv file should only have `survived` = 1.
4. The column `Pclass` should be one hot encoding.
5. Value of `Age` should be `young` if it is smaller than 65 and `old` otherwise.

```python
def titanic(original_file, modified_file):
    # BEGIN_YOUR_CODE
    df = pd.read_csv("titanic.csv")                     # read csv file
    df = df.drop(['Cabin', 'Embarked'], axis = 1)       # 1. remove Cabin, Embarked columns
    
                                                        # 2. erase row that contains missing value
    row2erase = []
    for colId, col in enumerate((df.columns)):
        for rowId, row in enumerate(pd.isna(df[col])):
            if row is True:
                row2erase.append(rowId)
                                                        # 3. erase survived=0 row
    for rowId, surv in enumerate(df['Survived']):
        if surv == 0:
            row2erase.append(rowId)
    
                                                        # 4. one hot encoding for Pclass
    df = pd.get_dummies(df, columns = ['Pclass'])
    
                                                        # 5. change Age value 
    def ageConverter(age):
        if age < 65:
            return 'young'
        return 'old'
    df['Age'] = df["Age"].apply(ageConverter)
    df = df.drop(row2erase, axis=0)
    
    # END_YOUR_CODE
    df.to_csv(modified_file, index=False)
```

### 6.a Employee Table [3 points]
Make employee table that has `name`, `salary` and `department_name` as columns. <br>
Note that each department has its own `department_id` and `department_name`.

```python
def emp_table(dep, emp1, emp2):
    # BEGIN_YOUR_CODE
    # read csv files
    emp1 = pd.read_csv(emp1)
    emp2 = pd.read_csv(emp2)
    dep = pd.read_csv(dep)
    
    def id2name(dep_id):
        return dep[dep['department_id'] == dep_id]["department_name"].values[0]
    # 
    df = pd.concat([emp1, emp2])
    df["department_id"] = df["department_id"].apply(id2name)
    df = df.rename(columns={"department_id":"department_name"})
    df = df.drop("employee_id", axis=1)
    # END_YOUR_CODE
    return df
```

### 6.b Highest Salary [3 points]
Find an employee with the highest salary of each department. <br>
The output must be a dictionary which contains `department_name` as a key and an employee's `name` as a value. <br>
You can use `emp_table` in 6.a.

```python
def lowest_avg(dep, emp1, emp2):
    # BEGIN_YOUR_CODE
    out = {}
    df = emp_table(dep, emp1, emp2)
    for department in pd.read_csv(dep)["department_name"]:

        filtered = df[df['department_name'] == department]["salary"].max()

        out[department] = filtered
    return out
    # END_YOUR_CODE
```
